```python
# from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import pandas as pd

hunt = pd.read_csv('Data\\Hunt\\huntfile.csv')
khar = pd.read_csv('Data\\Kharchenko\\kharchenkofile.csv')
dias = pd.read_csv('Data\\Dias\\diasfile.csv')
xmatch = pd.read_csv('Data\\Hunt\\xmatchfile.csv')

#replace the space by an underscore
dias['Cluster'] = dias['Cluster'].str.replace(' ', '_')
dias
```

```python
#crossmatch
dias_xm = xmatch.query('SourceCat == "Dias+02"').dropna(subset=['Sep']) #Get crossmatch table with seperation=NaN removed

dias_xm
```

```python
#remove duplicates
dias_xmnodup = dias_xm.drop_duplicates(subset=['ID'], inplace=False) 
dias_xmnodup
```

```python
#sort out globular clusters in hunt file
hunt_oc = hunt.loc[hunt['Type'] == 'o']
```

```python
#merge
diashunt = pd.merge(dias_xmnodup, hunt_oc, on= 'ID', how='inner')
diashunt
```

```python
#merge on names
dias_expanded = diashunt.assign(synonym=diashunt['AllNames'].str.split(',')).explode('synonym')
allnamemerge = pd.merge(dias_expanded, dias, left_on= 'synonym', right_on='Cluster', how='inner' )

allnamenodup = allnamemerge.drop_duplicates(subset=['synonym'], inplace=False) 
allnamenodup



```
