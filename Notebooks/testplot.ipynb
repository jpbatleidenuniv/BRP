```python
# from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)

fits_image_filename = 'J_A+A_633_A99_table1.dat.gz.fits'

hdul = fits.open(fits_image_filename)
hdr = hdul[0].header
hdr1 = hdul[1].header
data, cols = hdul[1].data, hdul[1].columns
print('Some metadata:')
for i, j in enumerate(cols.names):
    print(hdr1[f"TTYPE{i+1}"],"-->", hdr1.comments[f"TTYPE{i+1}"])

# print(hdr1)
```

```python
# Fig 3. of the paper
lowZ = (data['Z'] < 1800) & (data['Z'] > -1800)
fig, ax = plt.subplots()
ax.scatter(data[lowZ]['Rgc'], data[lowZ]['Z'], color='k', s=4)
plt.ylim(-1800, 1800)
plt.xlim(2000,15000)
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
NGC_6791 = data[data['cluster'] == 'NGC_6791']
# NGC_6791['Z']
plt.annotate('NGC 6791', [NGC_6791['Rgc'],NGC_6791['Z']], [0.6, 0.85], textcoords='axes fraction', arrowprops=dict(arrowstyle='->', shrinkB=5, connectionstyle = 'arc3, rad=0.3')) 
```
