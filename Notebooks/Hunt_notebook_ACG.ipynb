# Read catalogues and show columns

```python
hunt = pd.read_csv('../data/hunt_catalogue.csv')
kharchenko = pd.read_csv('../data/kharchenko_catalogue.csv')
hunt_xm = pd.read_csv('../data/hunt_crossmatch.csv')

print(hunt.columns)
print(kharchenko.columns)
print(hunt_xm.columns)
```

### Get only the objects crossmatched to Kharchenko's catalogue

```python
kharchenko_xm = hunt_xm.query('SourceCat == "Kharchenko+13"')
kharchenko_xm
```

### Get rid of the not crossmatched objects (Sep = Nan)

```python
kharchenko_xm = kharchenko_xm.dropna(subset = ['Sep'])
kharchenko_xm
```

### Show how many unique clusters in Hunt are crossmatched (unique ID)

```python
len(np.unique(kharchenko_xm['ID']))
```

### Count how many of these belong to the right type ('o')

```python
c = 0

for i in np.unique(kharchenko_xm['ID'])[1:2]:
    if i > 7166: continue
    type_object = hunt.query('ID == @i')['Type'][hunt.query('ID == @i').index[0]]
    if type_object == 'o': c = c + 1

print(c)
```

### Count how many of these belong to the right type in Kharchenko's catalogue (not 'g' -> globular cluster)

```python
kharchenko_names_oc = np.array(kharchenko.query('Type != "g"')['Name'],dtype = object)
```

```python
c = 0

for i in np.unique(kharchenko_xm['ID']):
    if i > 7166: continue
    type_object = hunt.query('ID == @i')['Type'][hunt.query('ID == @i').index[0]]
    if type_object == 'o': 
        for j in np.unique(kharchenko_xm.query('ID == @i')['NameLit'])[0:1]:
            if j in kharchenko_names_oc:
                c = c + 1

print(c)
```

### It can still be that the name we were checking was in the allNames column, check this
