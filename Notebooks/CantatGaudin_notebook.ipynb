```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import pandas as pd

cantat = pd.read_csv('Data\\CantatGaudin\\cantatgaudinfile.csv')
hunt = pd.read_csv('Data\\Hunt\\huntfile.csv')
xmatch = pd.read_csv('Data\\Hunt\\xmatchfile.csv').query('SourceCat == "Cantat-Gaudin+20"').dropna(subset='Sep')
```

```python
#Get open clusters from hunt
hunt_oc = hunt.query('Type == "o"')

#See how many crossmatched Cantat-Gaudin clusters were matched with hunt
huntxm = pd.merge(hunt_oc, xmatch, on='ID', how='inner').drop_duplicates('ID')
huntxm = huntxm.assign(synonym = huntxm['AllNames'].str.split(',')).explode('synonym').drop_duplicates('synonym') #Create AllNames column with synonyms of the OCs

#match the crossmatched Cantat-Gaudin clusters with the literature
cross = pd.merge(cantat, huntxm, left_on='Cluster', right_on='synonym', how='inner')
cantat_excl = pd.merge(cantat, huntxm, left_on='Cluster', right_on='synonym', how='left', indicator=True).query('_merge == "left_only"') #open clusters not crossmatched
```
