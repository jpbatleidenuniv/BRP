```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import pandas as pd

cantat = pd.read_csv('Data\\CantatGaudin\\cantatgaudinfile.csv')
hunt = pd.read_csv('Data\\Hunt\\huntfile.csv')
xmatch = pd.read_csv('Data\\Hunt\\xmatchfile.csv')

cantat_xm = xmatch.query('SourceCat=="Cantat-Gaudin+20"').drop_duplicates(subset=['ID'], inplace=False)
cantat.columns
```

```python
cantat_xm = xmatch.query('SourceCat == "Cantat-Gaudin+20"').dropna(subset=['Sep']) #Get crossmatch table with seperation=NaN removed
cantat_xm = cantat_xm.drop_duplicates(subset=['ID'], inplace=False)
```

```python
hunt_oc = hunt.query('Type=="o"')
cantathunt = pd.merge(cantat_xm, hunt_oc, on='ID', how='inner')
cantathunt
```

```python
cantatmerge = pd.merge(cantathunt, cantat, left_on='NameLit', right_on='Cluster', how='inner')
cantatmerge
```

```python
# cantat = cantat.query('')
cantat_expanded = cantathunt.assign(synonym=cantathunt['AllNames'].str.split(',')).explode('synonym')
cantat = cantat
allnamemerge = pd.merge(cantat_expanded, cantat, left_on='synonym', right_on='Cluster', how='inner')
allnamemerge
```
